**Moyenne arithmetique**

```r
moyVitess=mean(vitesse)
moyDistance=mean(distance)
cat("Moyenne de les vitesses  : ", moyVitess,"\n")
cat("Moyenne de les distances  : ", moyDistance)

```

**Calcule de la variance**

```r
varVitesse=var(vitesse)
varDistance=var(distance)
cat("Variance de la vitésse : ", varVitesse,"\n")
cat("Variance de la Distance : ", varDistance)
```

```r
covariance<-cov(vitesse,distance)
covariance
```
